```python
import tensorflow as tf

TRAIN_DIR = "../data/raw/NEU-DET/train"
VAL_DIR   = "../data/raw/NEU-DET/validation"

```

```python
import os

print(os.path.exists(TRAIN_DIR))
print(os.path.exists(VAL_DIR))
```

```python
IMG_SIZE = (128, 128)
BATCH_SIZE = 32

train_ds = tf.keras.utils.image_dataset_from_directory(
    TRAIN_DIR,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    shuffle=False
)

val_ds = tf.keras.utils.image_dataset_from_directory(
    VAL_DIR,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    shuffle=False
)
```

```python
class_names = train_ds.class_names
print("Classes:", class_names)
```

```python
class_names = val_ds.class_names
print("Classes:", class_names)
```

```python
def count_images_tf(dataset, class_names):
    counts = {cls: 0 for cls in class_names}
    
    for _, labels in dataset:
        for label in labels.numpy():
            counts[class_names[label]] += 1
            
    return counts

print("Train image counts:")
print(count_images_tf(train_ds, class_names))

print("\nValidation image counts:")
print(count_images_tf(val_ds, class_names))
```
